# Extração dos dados do Enade 2019

```python
#importar bibliotecas necessárias
import pandas as pd
import numpy as np
import zipfile
import requests
from io import BytesIO
import os
```

```python
# Criar um diretório para armazenar o conteúdo do ENADE
os.makedirs('./enade2019', exist_ok=True)
```

```python
# Define a URL
url = "https://download.inep.gov.br/microdados/microdados_enade_2019_LGPD.zip"

# Realiza o download do conteúdo

filebytes = BytesIO (
    requests.get(url, verify=False).content
)

#Extrair o conteúdo do zipfile
myzip = zipfile.ZipFile(filebytes)
myzip.extractall("./enade2019")
```
